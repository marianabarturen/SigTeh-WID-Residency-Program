## Two Portfolios

The following step was to build two portfolios. The **Clean Minus Emission** portfolio went LONG on the Clean Basket and SHORT on the Emission Basket. The **Emission minus Clean** went the other way around.

## The signals governing the portfolios

The signal governing the portfolio strategies were obtained from the sentiment analysis performed with stocktwits related to environmental issues. The response of the financial markets to this type of data is studied in __[Investors' Climate Sentiment and Financial Markets](https://www.researchgate.net/profile/Caterina-Santi-4/publication/344344260_Investors'_Climate_Sentiment_and_Financial_Markets/links/602e4b90299bf1cc26d2b651/)__, were it is observed that the markets respond to changes in Climate Sentiment.

Based on this information, **monthly delta signals** were constructed. 

Two versions of the signal were used: the **Raw** signal and the **Sign** signal were the sign of the the raw signal was taken. This leaves us with a signal formed with only -1, 0 or 1 values.

The Long-Short Strategy works as follows:
* if signal > 0, the it goes long, using the percentage of total assets indicated by tha absolute value of the signal
* if signal < 0, the it goes short, using the percentage of total assets indicated by tha absolute value of the signal
* if signal=0, no action

Values of 1 or -1 indicate to go fully long or short respectively.

Taking this into account, the Sign signal can be considered an extreme signal since it indicates to go fully long or short all the time.
